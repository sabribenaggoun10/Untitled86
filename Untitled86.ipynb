```python
def make_bar(value, max_value, width=30):
    bar_length = int(value[1]/max_value*width)
    if bar_length == 0:
        bar_length = 1
    return "#"*bar_length

def print_bar_chart(data,width=30,display=True):
    max_value = -1
    max_len = 0
    for _, v in data:
        if(type(_) is str):
            if len(_.strip())> max_len:
                max_len = len(_)
        if v > max_value:
            max_value =  v
    s = ""
    for element in data:
        bar = make_bar(element,max_value,width)
        if type(element[0]) is str:
            if display:
                print(f"{element[0]:>{max_len}}| {bar} {element[1]}")
            s+=f"{element[0]:>{max_len}}| {bar} {element[1]}\n"
        else:
            if display:
                print(f"{element[0]}| {bar} {element[1]}")
            s+=f"{element[0]}| {bar} {element[1]}\n"
    return s
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the visible half (e.g., two categories)
sizes = [30, 70]
labels = ['Category A', 'Category B']
colors = ['#ff9999', '#66b2ff']

# Add a dummy data point for the hidden half (sum of visible data)
# The sum of all values must represent a full circle.
total_sum = sum(sizes)
hidden_size = total_sum
sizes_full = sizes + [hidden_size]
# Add an empty label for the hidden part
labels_full = labels + ['']

# Create the plot
fig, ax = plt.subplots()
ax.pie(sizes_full,
       labels=labels_full,
       colors=colors + ['white'], # Use white color for the hidden part
       startangle=180,           # Start angle at 180 degrees (bottom of the circle)
       counterclock=False,       # Plot slices clockwise
       wedgeprops={'linewidth': 1, 'edgecolor': 'white'}) # Optional: customize wedge appearance

# Ensure the plot is a circle
ax.axis('equal')

# Adjust the y-axis limit to show only the top half
ax.set_ylim(0, 1.1) # Adjust the upper limit as needed for better spacing

plt.title('Half Pie Chart Example')
plt.show()
```

```python
# --- 1. SEPARATE THE DATA BY DIET ---
df_herbivores = df[df['diet'] == 'herbivore']
df_omnivores  = df[df['diet'] == 'omnivore']
df_carnivores = df[df['diet'] == 'carnivore']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

class LinearModel:
    def __init__(self, model_name=""):
        self.model_name = model_name

    def fit(self, x, y):
        x = pd.DataFrame(x)
        linear_model = LinearRegression().fit(x, y)
        y_pred = linear_model.predict(x)
        self.slope = linear_model.coef_[0]
        self.intercept = linear_model.intercept_
        self.rsquared = r2_score(y, y_pred)
        return self

    def predict(self, x):
        return self.slope * x + self.intercept

    def plot_model(self, x_min, x_max, color="black"):
        y_min = self.predict(x_min)
        y_max = self.predict(x_max)
        plt.plot([x_min, x_max], [y_min, y_max], color=color)

    def print_model_info(self):
        m = self.slope
        b = self.intercept
        rsquared = self.rsquared
        model_name = self.model_name
        print(f'LinearModel({model_name}):')
        print(f'Slope: {m}')
        print(f'Intercept: {b}')
        print(f'R-squared: {rsquared}')
```

```python
# --- 2. FIT A LINEAR MODEL FOR EACH DIET ---
# Herbivore Model
model_herbivore = LinearModel()
model_herbivore.fit(df_herbivores['log10_mass'], df_herbivores['sleep'])
print("--- Herbivores ---")
model_herbivore.print_model_info()
```

```python
# Omnivore Model
model_omnivore = LinearModel()
model_omnivore.fit(df_omnivores['log10_mass'], df_omnivores['sleep'])
print("\n--- Omnivores ---")
model_omnivore.print_model_info()

```

```python
# Carnivore Model
model_carnivore = LinearModel()
model_carnivore.fit(df_carnivores['log10_mass'], df_carnivores['sleep'])
print("\n--- Carnivores ---")
model_carnivore.print_model_info()
```

```python
# --- 3. VISUALIZE EVERYTHING TOGETHER ---
plt.figure(figsize=(10, 6))

# Plot the scatter points (Real Data)
plt.scatter(df_herbivores['log10_mass'], df_herbivores['sleep'], color='green', label='Herbivores', marker='o')
plt.scatter(df_omnivores['log10_mass'],  df_omnivores['sleep'],  color='blue',  label='Omnivores',  marker='s')
plt.scatter(df_carnivores['log10_mass'], df_carnivores['sleep'], color='red',   label='Carnivores', marker='^')

# Get the min and max x-values to draw the lines across the whole chart
x_min = df['log10_mass'].min()
x_max = df['log10_mass'].max()

# Plot the lines (The AI Models)
model_herbivore.plot_model(x_min, x_max, color='green')
model_omnivore.plot_model(x_min, x_max, color='blue')
model_carnivore.plot_model(x_min, x_max, color='red')

# Format the chart to be readable (Honest Axes & Clean Labels)
plt.title('How Body Mass and Diet Affect Animal Sleep', fontsize=14)
plt.xlabel('Body Mass (kg) - Log Scale', fontsize=12)
plt.ylabel('Hours of sleep per day', fontsize=12)

# Apply the custom log scale labels provided in the starter code
xticks = [-2, -1, 0, 1, 2, 3, 4]
xtick_labels = [0.01, 0.1, 1, 10, 100, 1000, 10000]
plt.xticks(xticks, xtick_labels)

plt.legend()
plt.show()
```
